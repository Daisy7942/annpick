```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# MySQL 연결 설정 함수
def create_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("Connection to MySQL DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")
    return connection

# recommendation_phrase 필드를 업데이트하는 함수
def update_recommendation_phrases(connection, update_data):
    cursor = connection.cursor()

    for index, row in update_data.iterrows():
        genre_tag = row['genre + tag']
        new_recommendation_phrase = row['recommendation_phrase']

        # 기존 recommendation_phrase 필드가 genre + tag 문자열과 일치하는 경우, 추천 문구 업데이트
        update_query = """
        UPDATE RecommendationCluster
        SET recommendation_phrase = %s
        WHERE recommendation_phrase = %s
        """
        try:
            cursor.execute(update_query, (new_recommendation_phrase, genre_tag))
            connection.commit()
            print(f"Updated recommendation phrase for {genre_tag} to: {new_recommendation_phrase}")
        except Error as e:
            print(f"Error occurred while updating {genre_tag}: {e}")

    cursor.close()

# MySQL 연결
connection = create_connection("RDS주소", "username", "password", "table")

# 엑셀 파일 로드
file_path = '/content/recommend_phrase.csv'
recommend_phrase_df = pd.read_csv(file_path)

# recommendation_phrase 필드를 업데이트
update_recommendation_phrases(connection, recommend_phrase_df)
```
